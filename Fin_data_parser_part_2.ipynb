Convert messy currency strings to floats.
- Handles commas, currency symbols, spaces
- Handles negatives in parentheses, e.g., (123.45) -> -123.45
- Leaves NaN as NaN


```python
def to_amount(series):
 
    def parse_one(x):
        if pd.isna(x):
            return np.nan
        s = str(x).strip()
        # Check for parentheses negative format
        neg = s.startswith("(") and s.endswith(")")
        if neg:
            s = s[1:-1]
        # Keep only digits, dot, minus (remove currency symbols/commas etc.)
        s = re.sub(r"[^\d.\-]", "", s)
        if s in {"", ".", "-"}:
            return np.nan
        try:
            v = float(s)
        except Exception:
            return np.nan
        return -v if neg else v

    return series.apply(parse_one)


BANK_PATH    = "KH_Bank.XLSX"
LEDGER_PATH  = "Customer_Ledger_Entries_FULL.xlsx"
BANK_SHEET   = "Sheet1"
LEDGER_SHEET = "Customer Ledger Entries"

# Confirmed column choices:
BANK_AMOUNT_COL = "Statement.Entry.Amount.Value"
BANK_DESC_COL   = "Statement.Account.Identification.IBAN"
LEDGER_TGT_COL  = "Original Amount"
LEDGER_REF_COL  = "Document No."

print("Loaded:")
print(f"  Bank   -> shape={df.shape}, sheet='{BANK_SHEET}'")
print(f"  Ledger -> shape={df2.shape}, sheet='{LEDGER_SHEET}'")
print("\n[Preview] Bank (first 5 rows):")
df.head(5)
```

```python
import pandas as pd
import numpy as np
from itertools import combinations
from math import gcd
from functools import reduce


BANK_ID_COL       = "tx_id"
BANK_AMOUNT_COL   = "amount"
LEDGER_ID_COL     = "target_id"
LEDGER_AMOUNT_COL = "target_amount"


try:
    bank_unmatched 
    ledger_unmatched
except NameError:
    bank_unmatched   = bank.copy()
    ledger_unmatched = ledger.copy()


bank_unmatched[BANK_AMOUNT_COL]     = pd.to_numeric(bank_unmatched[BANK_AMOUNT_COL], errors="coerce")
ledger_unmatched[LEDGER_AMOUNT_COL] = pd.to_numeric(ledger_unmatched[LEDGER_AMOUNT_COL], errors="coerce").abs()


def prepare_ledger_pool(df, amount_col=LEDGER_AMOUNT_COL):
    x = df.copy()
    x["amt_abs"] = pd.to_numeric(x[amount_col], errors="coerce").abs()
    x = x[x["amt_abs"].notna()]
    x["cents"] = (x["amt_abs"] * 100).round().astype(np.int64)
    return x

ledger_pool = prepare_ledger_pool(ledger_unmatched, LEDGER_AMOUNT_COL)


def pick_candidates_fast(pool, target_c, max_k=24):
    view = pool[pool["cents"] <= target_c]
    if view.empty:
        return view
    if len(view) <= max_k:
        return view
    diffs = np.abs(view["cents"].to_numpy() - target_c)
    idx = np.argpartition(diffs, max_k - 1)[:max_k]
    return view.iloc[idx]


def gcd_reduce(arr):
    return reduce(gcd, arr) if len(arr) else 0


def mitm_subset_sum(cand_df, target_c):
    # items: [(row_index, cents), ...]
    items = list(zip(cand_df.index.tolist(), cand_df["cents"].tolist()))
    n = len(items)
    if n == 0:
        return None


    mid = n // 2
    L, R = items[:mid], items[mid:]


    sums_L = {0: []}  
    for r in range(1, len(L) + 1):
        for combo in combinations(L, r):
            s = sum(w for _, w in combo)
            if s not in sums_L or len(combo) < len(sums_L[s]):
                sums_L[s] = [idx for idx, _ in combo]


    if target_c in sums_L:
        return sums_L[target_c]


    for r in range(1, len(R) + 1):
        for combo in combinations(R, r):
            s = sum(w for _, w in combo)
            if s == target_c:
                return [idx for idx, _ in combo]
            need = target_c - s
            if need in sums_L:
                return sums_L[need] + [idx for idx, _ in combo]

    return None


MAX_K = 24  

matches = []
rows = len(bank_unmatched)

for i, (_, brow) in enumerate(bank_unmatched.iterrows(), 1):
    target = pd.to_numeric(brow[BANK_AMOUNT_COL], errors="coerce")
    if pd.isna(target) or target == 0:
        continue
    target_c = int(round(abs(float(target)) * 100))


    cand = pick_candidates_fast(ledger_pool, target_c, max_k=MAX_K)
    if cand.empty:
        continue


    cents_arr = cand["cents"].to_numpy(np.int64)
    if cents_arr.min(initial=target_c+1) > target_c: 
        continue
    if cents_arr.sum() < target_c:  
        continue
    g = gcd_reduce(cents_arr.tolist())
    if g and (target_c % g != 0):               
        continue


    idxs = mitm_subset_sum(cand, target_c)

    if idxs:
        block = ledger_pool.loc[idxs, [LEDGER_ID_COL, LEDGER_AMOUNT_COL]]
        matches.append({
            "bank_tx_id":  brow[BANK_ID_COL],
            "bank_amount": round(float(target), 2),
            "ledger_ids":  block[LEDGER_ID_COL].tolist(),
            "ledger_amts": block[LEDGER_AMOUNT_COL].round(2).tolist(),
            "sum_ledger":  round(float(block[LEDGER_AMOUNT_COL].sum()), 2),
            "n_items":     len(block),
            "method":      "bruteforce_mitm"
        })

        ledger_pool = ledger_pool.drop(index=idxs)


    if i % 100 == 0:
        print(f"Processed {i}/{rows} bank rows; matches so far: {len(matches)}")

matches_df = pd.DataFrame(matches)
print("Group matches found:", len(matches_df))
matches_df.head()
```

## Task 2.3: Performance Analysis

```python
import time
import numpy as np
import pandas as pd
from itertools import combinations
from math import gcd
from functools import reduce


BANK_ID_COL       = "tx_id"
BANK_AMOUNT_COL   = "amount"
LEDGER_ID_COL     = "target_id"
LEDGER_AMOUNT_COL = "target_amount"


bank_pool = globals().get("bank_unmatched_after_2_1", globals().get("bank_unmatched", bank.copy()))
ledger_pool = globals().get("ledger_unmatched_after_2_1", globals().get("ledger_unmatched", ledger.copy()))


bank_pool = bank_pool.copy()
ledger_pool = ledger_pool.copy()
bank_pool[BANK_AMOUNT_COL]     = pd.to_numeric(bank_pool[BANK_AMOUNT_COL], errors="coerce")
ledger_pool[LEDGER_AMOUNT_COL] = pd.to_numeric(ledger_pool[LEDGER_AMOUNT_COL], errors="coerce").abs()


ledger_sorted = (
    ledger_pool
      .assign(cents=(ledger_pool[LEDGER_AMOUNT_COL] * 100).round().astype("Int64"))
      .dropna(subset=["cents"])
      .sort_values("cents")
      .reset_index()              
      .rename(columns={"index": "orig_idx"})
)

cents_arr = ledger_sorted["cents"].to_numpy(np.int64)
N_LEDGER  = len(ledger_sorted)

def fresh_availability_mask():
    
    return np.ones(N_LEDGER, dtype=bool)

def gcd_reduce(arr):
    return reduce(gcd, arr) if len(arr) else 0
```

```python
def pick_window_by_amount(target_c, avail_mask, abs_tol_c=500, rel_tol=0.02, max_k=24):

    tol = max(abs_tol_c, int(round(target_c * rel_tol)))
    lo_c, hi_c = max(0, target_c - tol), target_c + tol

    lo = np.searchsorted(cents_arr, lo_c, side="left")
    hi = np.searchsorted(cents_arr, hi_c, side="right")
    if lo >= hi:
        return ledger_sorted.iloc[[]]

    window_pos = np.arange(lo, hi)
    mask = avail_mask[lo:hi]
    pos = window_pos[mask]
    if pos.size == 0:
        return ledger_sorted.iloc[[]]

    diffs = np.abs(cents_arr[pos] - target_c)
    k = min(max_k, pos.size)
    sel_local = np.argpartition(diffs, k - 1)[:k]
    sel_pos = pos[sel_local]
    return ledger_sorted.iloc[sel_pos]


def mitm_subset_sum(cand_df, target_c):

    items = list(zip(cand_df["orig_idx"].tolist(), cand_df["cents"].tolist()))
    n = len(items)
    if n == 0:
        return None
    mid = n // 2
    L, R = items[:mid], items[mid:]

    sums_L = {0: []}
    for r in range(1, len(L) + 1):
        for combo in combinations(L, r):
            s = sum(w for _, w in combo)
            cur = [idx for idx, _ in combo]
            if (s not in sums_L) or (len(cur) < len(sums_L[s])):
                sums_L[s] = cur

    if target_c in sums_L:
        return sums_L[target_c]

    for r in range(1, len(R) + 1):
        for combo in combinations(R, r):
            s = sum(w for _, w in combo)
            if s == target_c:
                return [idx for idx, _ in combo]
            need = target_c - s
            if need in sums_L:
                return sums_L[need] + [idx for idx, _ in combo]
    return None


def dp_subset_sum(cand_df, target_c):
    items = list(zip(cand_df["orig_idx"].tolist(), cand_df["cents"].tolist()))
    if not items:
        return None
    parent = {0: None} 

    for idx, w in items:
        for s in list(parent.keys()):
            ns = s + w
            if ns > target_c or ns in parent:
                continue
            parent[ns] = (s, idx)
            if ns == target_c:
                sel = []
                cur = ns
                while cur != 0:
                    prev, used_idx = parent[cur]
                    sel.append(used_idx)
                    cur = prev
                sel.reverse()
                return sel
    return None
```

```python
def reconcile_one_pass(bank_df,
                       method="mitm",
                       max_k=24,
                       abs_tol_c=500,
                       rel_tol=0.02,
                       time_budget_sec=None):
    """
    Runs a single pass over bank_df using the chosen solver.
    Returns: (matches_df, stats_dict)
    """
    assert method in ("mitm", "dp")
    avail = fresh_availability_mask()
    matches = []
    total_candidates = 0
    started = time.perf_counter()

    for _, brow in bank_df.iterrows():
        if time_budget_sec and (time.perf_counter() - started) > time_budget_sec:
            break

        a = pd.to_numeric(brow[BANK_AMOUNT_COL], errors="coerce")
        if pd.isna(a) or a == 0:
            continue
        target_c = int(round(abs(float(a)) * 100))

        cand = pick_window_by_amount(target_c, avail, abs_tol_c=abs_tol_c, rel_tol=rel_tol, max_k=max_k)
        if cand.empty:
            continue
        cents_vec = cand["cents"].to_numpy(np.int64)
        total_candidates += len(cand)

        
        if cents_vec.min(initial=target_c+1) > target_c:
            continue
        if cents_vec.sum() < target_c:
            continue
        g = gcd_reduce(cents_vec.tolist())
        if g and (target_c % g != 0):
            continue

        if method == "mitm":
            chosen = mitm_subset_sum(cand, target_c)
        else:
            chosen = dp_subset_sum(cand, target_c)

        if chosen:
            block = ledger_pool.loc[chosen, [LEDGER_ID_COL, LEDGER_AMOUNT_COL]]
            matches.append({
                "bank_tx_id":  brow[BANK_ID_COL],
                "bank_amount": round(float(a), 2),
                "ledger_ids":  block[LEDGER_ID_COL].tolist(),
                "ledger_amts": block[LEDGER_AMOUNT_COL].round(2).tolist(),
                "sum_ledger":  round(float(block[LEDGER_AMOUNT_COL].sum()), 2),
                "n_items":     len(block),
                "method":      method,
            })
            
            pos = pd.Series(np.arange(N_LEDGER), index=ledger_sorted["orig_idx"]).loc[chosen].to_numpy()
            avail[pos] = False

    took = time.perf_counter() - started
    matches_df = pd.DataFrame(matches)
    stats = {
        "rows_processed": len(bank_df),
        "matches_found": len(matches_df),
        "avg_candidates": (total_candidates / max(1, len(bank_df))),
        "seconds": took,
        "ms_per_row": (took / max(1, len(bank_df))) * 1000.0,
        "method": method,
        "max_k": max_k,
        "abs_tol_c": abs_tol_c,
        "rel_tol": rel_tol,
    }
    return matches_df, stats
```

```python
def run_benchmark(sizes=(100, 250, 500, 750, 1000),
                  method="mitm",
                  repeats=2,
                  max_k=24,
                  abs_tol_c=500,
                  rel_tol=0.02,
                  random_state=42,
                  time_budget_sec=None):
    """
    Measure execution time for increasing bank sizes.
    """
    rng = np.random.default_rng(random_state)
    results = []
    for n in sizes:
        
        n_eff = min(n, len(bank_pool))
        sample_idx = rng.choice(bank_pool.index.to_numpy(), size=n_eff, replace=False)
        bank_chunk = bank_pool.loc[sample_idx].reset_index(drop=True)

        for r in range(repeats):
            _, stats = reconcile_one_pass(
                bank_chunk,
                method=method,
                max_k=max_k,
                abs_tol_c=abs_tol_c,
                rel_tol=rel_tol,
                time_budget_sec=time_budget_sec,
            )
            stats.update({"bank_rows": n_eff, "repeat": r+1})
            results.append(stats)
            print(f"{method.upper()} | N={n_eff:4d} | rep {r+1}/{repeats} | "
                  f"{stats['seconds']:.2f}s total, {stats['ms_per_row']:.2f} ms/row, "
                  f"matches={stats['matches_found']}, avgK={stats['avg_candidates']:.1f}")

    results_df = pd.DataFrame(results)
    display(results_df)
    results_df.to_csv(f"task_2_3_perf_{method}.csv", index=False)
    print(f"Saved results â†’ task_2_3_perf_{method}.csv")
    return results_df
```

```python
import pandas as pd
import numpy as np
from itertools import combinations
from math import gcd
from functools import reduce


BANK_ID_COL       = "tx_id"
BANK_AMOUNT_COL   = "amount"
LEDGER_ID_COL     = "target_id"
LEDGER_AMOUNT_COL = "target_amount"


try:
    bank_pool   = bank_unmatched.copy()
    ledger_pool = ledger_unmatched.copy()
except NameError:
    bank_pool   = bank.copy()
    ledger_pool = ledger.copy()


bank_pool[BANK_AMOUNT_COL]     = pd.to_numeric(bank_pool[BANK_AMOUNT_COL], errors="coerce")
ledger_pool[LEDGER_AMOUNT_COL] = pd.to_numeric(ledger_pool[LEDGER_AMOUNT_COL], errors="coerce").abs()


ledger_sorted = (
    ledger_pool
      .assign(cents=(ledger_pool[LEDGER_AMOUNT_COL] * 100).round().astype("Int64"))
      .dropna(subset=["cents"])
      .sort_values("cents")
      .reset_index()   
      .rename(columns={"index":"orig_idx"})
)
cents_arr = ledger_sorted["cents"].to_numpy(np.int64)
N = len(ledger_sorted)
avail = np.ones(N, dtype=bool)


pos_by_orig = pd.Series(np.arange(N), index=ledger_sorted["orig_idx"].values)
```

```python
def pick_window_by_amount(target_c, abs_tol_c=500, rel_tol=0.02, max_k=24):

    tol = max(abs_tol_c, int(round(target_c * rel_tol)))
    lo_c, hi_c = max(0, target_c - tol), target_c + tol


    lo = np.searchsorted(cents_arr, lo_c, side="left")
    hi = np.searchsorted(cents_arr, hi_c, side="right")
    if lo >= hi:
        return ledger_sorted.iloc[[]]


    window_pos = np.arange(lo, hi)
    mask = avail[lo:hi]
    pos = window_pos[mask]
    if pos.size == 0:
        return ledger_sorted.iloc[[]]


    diffs = np.abs(cents_arr[pos] - target_c)
    k = min(max_k, pos.size)
    sel_local = np.argpartition(diffs, k-1)[:k]
    sel_pos = pos[sel_local]
    return ledger_sorted.iloc[sel_pos]

def gcd_reduce(arr):
    return reduce(gcd, arr) if len(arr) else 0

def mitm_subset_sum(cand_df, target_c):

    items = list(zip(cand_df["orig_idx"].tolist(), cand_df["cents"].tolist()))
    n = len(items)
    if n == 0:
        return None

    mid = n // 2
    L, R = items[:mid], items[mid:]


    sums_L = {0: []}
    for r in range(1, len(L) + 1):
        for combo in combinations(L, r):
            s = sum(w for _, w in combo)

            cur = [idx for idx, _ in combo]
            if (s not in sums_L) or (len(cur) < len(sums_L[s])):
                sums_L[s] = cur

    if target_c in sums_L:
        return sums_L[target_c]


    for r in range(1, len(R) + 1):
        for combo in combinations(R, r):
            s = sum(w for _, w in combo)
            if s == target_c:
                return [idx for idx, _ in combo]
            need = target_c - s
            if need in sums_L:
                return sums_L[need] + [idx for idx, _ in combo]

    return None
```

```python
import numpy as np, pandas as pd
from itertools import combinations
from math import gcd
from functools import reduce

BANK_ID_COL       = "tx_id"
BANK_AMOUNT_COL   = "amount"
LEDGER_ID_COL     = "target_id"
LEDGER_AMOUNT_COL = "target_amount"


bank_pool   = globals().get("bank_unmatched", bank).copy()
ledger_pool = globals().get("ledger_unmatched", ledger).copy()


bank_pool[BANK_AMOUNT_COL]     = pd.to_numeric(bank_pool[BANK_AMOUNT_COL], errors="coerce")
ledger_pool[LEDGER_AMOUNT_COL] = pd.to_numeric(ledger_pool[LEDGER_AMOUNT_COL], errors="coerce").abs()


ledger_sorted = (
    ledger_pool
      .assign(cents=(ledger_pool[LEDGER_AMOUNT_COL]*100).round().astype("Int64"))
      .dropna(subset=["cents"])
      .sort_values("cents")
      .reset_index()
      .rename(columns={"index":"orig_idx"})
)
cents_arr  = ledger_sorted["cents"].to_numpy(np.int64)
pos_by_orig = pd.Series(np.arange(len(ledger_sorted)), index=ledger_sorted["orig_idx"].values)
avail = np.ones(len(ledger_sorted), dtype=bool)


def gcd_reduce(arr): return reduce(gcd, arr) if len(arr) else 0

def pick_window(target_c, abs_tol_c, rel_tol, max_k):
    tol = max(abs_tol_c, int(round(target_c*rel_tol)))
    lo_c, hi_c = max(0, target_c - tol), target_c + tol
    lo = np.searchsorted(cents_arr, lo_c, side="left")
    hi = np.searchsorted(cents_arr, hi_c, side="right")
    if lo >= hi: return ledger_sorted.iloc[[]]
    window_pos = np.arange(lo, hi)
    mask = avail[lo:hi]
    pos = window_pos[mask]
    if pos.size == 0: return ledger_sorted.iloc[[]]
    diffs = np.abs(cents_arr[pos] - target_c)
    k = min(max_k, pos.size)
    sel = pos[np.argpartition(diffs, k-1)[:k]]
    return ledger_sorted.iloc[sel]

def one_sum(cand, target_c):
    hit = cand.loc[cand["cents"].eq(target_c)]
    return [int(hit.iloc[0]["orig_idx"])] if len(hit) else None

def two_sum(cand, target_c):
    seen = {}
    for _, r in cand.iterrows():
        c = int(r["cents"]); i = int(r["orig_idx"])
        need = target_c - c
        if need in seen: return [seen[need], i]
        seen[c] = i
    return None


def mitm_subset_sum(cand, target_c):
    items = list(zip(cand["orig_idx"].tolist(), cand["cents"].tolist()))
    n = len(items)
    if n == 0: return None
    mid = n//2; L, R = items[:mid], items[mid:]
    sumsL = {0: []}
    for r in range(1, len(L)+1):
        for combo in combinations(L, r):
            s = sum(w for _, w in combo)
            cur = [idx for idx, _ in combo]
            if (s not in sumsL) or (len(cur) < len(sumsL[s])): sumsL[s] = cur
    if target_c in sumsL: return sumsL[target_c]
    for r in range(1, len(R)+1):
        for combo in combinations(R, r):
            s = sum(w for _, w in combo)
            if s == target_c: return [idx for idx, _ in combo]
            need = target_c - s
            if need in sumsL: return sumsL[need] + [idx for idx, _ in combo]
    return None


def dp_subset_sum(cand, target_c):
    items = list(zip(cand["orig_idx"].tolist(), cand["cents"].tolist()))
    if not items: return None
    parent = {0: None}
    for idx, w in items:
        for s in list(parent.keys()):
            ns = s + w
            if ns > target_c or ns in parent: continue
            parent[ns] = (s, idx)
            if ns == target_c:
                sel = []
                cur = ns
                while cur != 0:
                    prev, used_idx = parent[cur]
                    sel.append(used_idx); cur = prev
                return list(reversed(sel))
    return None


PHASES = [
    (500,  0.02, 16, ("1","2","mitm")),          # Â±5, 2%, K=16
    (1000, 0.03, 24, ("1","2","mitm")),          # Â±10, 3%, K=24
    (2000, 0.05, 32, ("mitm","dp")),             # Â±20, 5%, K=32
    (3000, 0.07, 40, ("dp",)),                    # Â±30, 7%, K=40
]

def find_subset_adaptive(target_c):
    for abs_tol_c, rel_tol, max_k, solvers in PHASES:
        cand = pick_window(target_c, abs_tol_c, rel_tol, max_k)
        if cand.empty: continue
        cents_vec = cand["cents"].to_numpy(np.int64)
        if cents_vec.min(initial=target_c+1) > target_c: continue
        if cents_vec.sum() < target_c: continue
        g = gcd_reduce(cents_vec.tolist())
        if g and (target_c % g != 0): continue

        for s in solvers:
            if s == "1":
                pick = one_sum(cand, target_c)
            elif s == "2":
                pick = two_sum(cand, target_c)
            elif s == "mitm":
                pick = mitm_subset_sum(cand, target_c)
            elif s == "dp":
                pick = dp_subset_sum(cand, target_c)
            else:
                pick = None
            if pick: return pick
    return None


matches = []
rows = len(bank_pool)
for i, (_, brow) in enumerate(bank_pool.iterrows(), 1):
    a = pd.to_numeric(brow[BANK_AMOUNT_COL], errors="coerce")
    if pd.isna(a) or a == 0: continue
    target_c = int(round(abs(float(a))*100))

    chosen = find_subset_adaptive(target_c)
    if chosen:
        block = ledger_pool.loc[chosen, [LEDGER_ID_COL, LEDGER_AMOUNT_COL]]
        matches.append({
            "bank_tx_id":  brow[BANK_ID_COL],
            "bank_amount": round(float(a), 2),
            "ledger_ids":  block[LEDGER_ID_COL].tolist(),
            "ledger_amts": block[LEDGER_AMOUNT_COL].round(2).tolist(),
            "sum_ledger":  round(float(block[LEDGER_AMOUNT_COL].sum()), 2),
            "n_items":     len(block),
            "method":      "adaptive"
        })

        avail[pos_by_orig[chosen].to_numpy()] = False

    if i % 100 == 0:
        print(f"processed {i}/{rows}; matches so far: {len(matches)}")

matches_df = pd.DataFrame(matches)
print("Adaptive group matches found:", len(matches_df))
matches_df.head()
```

```python
BUCKET = 100.0
bank_ml["amt_bucket"]   = (bank_ml["amt_abs"]   // BUCKET).astype("Int64")
ledger_ml["amt_bucket"] = (ledger_ml["amt_abs"] // BUCKET).astype("Int64")

b_small = bank_ml[[BANK_ID_COL,"amt_abs","amt_bucket","desc_clean"]].dropna(subset=["amt_abs"])
l_small = ledger_ml[[LEDGER_ID_COL,"amt_abs","amt_bucket","desc_clean"]].dropna(subset=["amt_abs"])

pairs = b_small.merge(l_small, on="amt_bucket", how="inner", suffixes=("_bank","_led"))


AMT_ABS_TOL = 5.0   
AMT_REL_TOL = 0.02    

pairs["abs_diff"] = (pairs["amt_abs_bank"] - pairs["amt_abs_led"]).abs()
pairs["rel_diff"] = pairs["abs_diff"] / pairs["amt_abs_bank"].clip(lower=1e-6)
pairs = pairs[(pairs["abs_diff"] <= AMT_ABS_TOL) | (pairs["rel_diff"] <= AMT_REL_TOL)].copy()
```

```python
def jaccard(a, b):
    sa = set(a.split()) if a else set()
    sb = set(b.split()) if b else set()
    if not sa and not sb: return 0.0
    return len(sa & sb) / max(1, len(sa | sb))

def seq_ratio(a, b):    
    return SequenceMatcher(a=a or "", b=b or "").ratio()

def trigram_bag(s):
    s = f"  {s}  "        
    return [s[i:i+3] for i in range(max(0, len(s)-2))] if s else []

def trigram_cosine(a, b):
    A = trigram_bag(a); B = trigram_bag(b)
    if not A and not B: return 0.0
    from collections import Counter
    ca, cb = Counter(A), Counter(B)
    dot = sum(ca[g]*cb.get(g,0) for g in ca)
    na = np.sqrt(sum(v*v for v in ca.values())); nb = np.sqrt(sum(v*v for v in cb.values()))
    return float(dot)/(na*nb) if na and nb else 0.0

def num_overlap(a, b):
    A = set(re.findall(r"\d+", a or "")); B = set(re.findall(r"\d+", b or ""))
    return len(A & B)
```

```python
pairs["seq_ratio"]   = [seq_ratio(a,b)   for a,b in zip(pairs["desc_clean_bank"], pairs["desc_clean_led"])]
pairs["jaccard"]     = [jaccard(a,b)     for a,b in zip(pairs["desc_clean_bank"], pairs["desc_clean_led"])]
pairs["tri_cosine"]  = [trigram_cosine(a,b) for a,b in zip(pairs["desc_clean_bank"], pairs["desc_clean_led"])]
pairs["num_overlap"] = [num_overlap(a,b) for a,b in zip(pairs["desc_clean_bank"], pairs["desc_clean_led"])]
```

```python

amt_s_abs = 1 - (pairs["abs_diff"] / AMT_ABS_TOL).clip(upper=1)
amt_s_rel = 1 - (pairs["rel_diff"] / AMT_REL_TOL).clip(upper=1)
pairs["amount_score"] = 0.5*amt_s_abs + 0.5*amt_s_rel


text_base = 0.5*pairs["seq_ratio"] + 0.3*pairs["jaccard"] + 0.2*pairs["tri_cosine"]
pairs["text_score"] = (text_base + (pairs["num_overlap"]>0)*0.05).clip(upper=1)


W_AMT, W_TEXT = 0.6, 0.4        
pairs["fuzzy_score"] = (W_AMT*pairs["amount_score"] + W_TEXT*pairs["text_score"]).round(4)
```

```python
import time, numpy as np, pandas as pd, re
from itertools import combinations
from math import gcd
from functools import reduce
from difflib import SequenceMatcher


BANK_ID_COL, BANK_AMOUNT_COL, BANK_DESC_COL = "tx_id", "amount", "description"
LEDGER_ID_COL, LEDGER_AMOUNT_COL, LEDGER_DESC_COL = "target_id", "target_amount", "reference_id"


bank_base   = globals().get("bank_unmatched", bank).copy()
ledger_base = globals().get("ledger_unmatched", ledger).copy()


bank_base[BANK_AMOUNT_COL]     = pd.to_numeric(bank_base[BANK_AMOUNT_COL], errors="coerce")
ledger_base[LEDGER_AMOUNT_COL] = pd.to_numeric(ledger_base[LEDGER_AMOUNT_COL], errors="coerce").abs()

def build_sorted_ledger(ledger_df):
    Ls = (
        ledger_df
          .assign(cents=(ledger_df[LEDGER_AMOUNT_COL]*100).round().astype("Int64"))
          .dropna(subset=["cents"])
          .sort_values("cents")
          .reset_index()
          .rename(columns={"index":"orig_idx"})
    )
    cents = Ls["cents"].to_numpy(np.int64)
    pos_by_orig = pd.Series(np.arange(len(Ls)), index=Ls["orig_idx"].values)
    return Ls, cents, pos_by_orig

def fresh_avail(n): return np.ones(n, dtype=bool)
def gcd_reduce(arr): return reduce(gcd, arr) if len(arr) else 0

def pick_window_by_amount(target_c, cents_arr, ledger_sorted, avail_mask,
                          abs_tol_c=500, rel_tol=0.02, max_k=24):
    tol = max(abs_tol_c, int(round(target_c*rel_tol)))
    lo_c, hi_c = max(0, target_c - tol), target_c + tol
    lo = np.searchsorted(cents_arr, lo_c, side="left")
    hi = np.searchsorted(cents_arr, hi_c, side="right")
    if lo >= hi: return ledger_sorted.iloc[[]]
    window_pos = np.arange(lo, hi)
    pos = window_pos[avail_mask[lo:hi]]
    if pos.size == 0: return ledger_sorted.iloc[[]]
    diffs = np.abs(cents_arr[pos] - target_c)
    k = min(max_k, pos.size)
    sel = pos[np.argpartition(diffs, k-1)[:k]]
    return ledger_sorted.iloc[sel]
```

```python
def one_sum(cand, target_c):
    hit = cand.loc[cand["cents"].eq(target_c)]
    return [int(hit.iloc[0]["orig_idx"])] if len(hit) else None

def two_sum(cand, target_c):
    seen = {}
    for _, r in cand.iterrows():
        c = int(r["cents"]); i = int(r["orig_idx"])
        need = target_c - c
        if need in seen: return [seen[need], i]
        seen[c] = i
    return None

def mitm_subset_sum(cand, target_c):
    items = list(zip(cand["orig_idx"].tolist(), cand["cents"].tolist()))
    n = len(items)
    if n == 0: return None
    mid = n//2; L, R = items[:mid], items[mid:]
    sumsL = {0: []}
    for r in range(1, len(L)+1):
        for combo in combinations(L, r):
            s = sum(w for _, w in combo)
            cur = [idx for idx, _ in combo]
            if (s not in sumsL) or (len(cur) < len(sumsL[s])): sumsL[s] = cur
    if target_c in sumsL: return sumsL[target_c]
    for r in range(1, len(R)+1):
        for combo in combinations(R, r):
            s = sum(w for _, w in combo)
            if s == target_c: return [idx for idx, _ in combo]
            need = target_c - s
            if need in sumsL: return sumsL[need] + [idx for idx, _ in combo]
    return None

def dp_subset_sum(cand, target_c):
    items = list(zip(cand["orig_idx"].tolist(), cand["cents"].tolist()))
    if not items: return None
    parent = {0: None}
    for idx, w in items:
        for s in list(parent.keys()):
            ns = s + w
            if ns > target_c or ns in parent: continue
            parent[ns] = (s, idx)
            if ns == target_c:
                sel, cur = [], ns
                while cur != 0:
                    prev, used = parent[cur]; sel.append(used); cur = prev
                return list(reversed(sel))
    return None


def ga_subset_sum(cand, target_c, max_items=6, pop_size=64, generations=120,
                  mutation_rate=0.02, elite_frac=0.05, tournament_k=3,
                  time_budget_sec=0.3, seed=42):
    import time as _t
    rng = np.random.default_rng(seed)
    w   = cand["cents"].to_numpy(np.int64)
    idx = cand["orig_idx"].to_numpy(np.int64)
    M   = len(w)
    if M == 0: return None
    start = _t.perf_counter()

    pop = rng.integers(0, 2, size=(pop_size, M), dtype=np.int8)
    if max_items:
        sums = pop.sum(axis=1)
        for i, s in enumerate(sums):
            if s > max_items:
                ones = np.where(pop[i]==1)[0]
                drop = rng.choice(ones, size=s-max_items, replace=False)
                pop[i, drop] = 0

    for _ in range(generations):
        sums = pop @ w
        err  = np.abs(sums - target_c)
        z = np.where(err == 0)[0]
        if z.size:
            sel_cols = np.where(pop[z[0]] == 1)[0]
            return idx[sel_cols].tolist()
        if _t.perf_counter() - start > time_budget_sec: break


        cand_ix = rng.integers(0, pop_size, size=(pop_size, tournament_k))
        winners = cand_ix[np.arange(pop_size), np.argmin(err[cand_ix], axis=1)]
        parents = pop[winners]


        half = pop_size//2
        p1, p2 = parents[:half], parents[half:]
        mask = rng.integers(0, 2, size=p1.shape, dtype=np.int8)
        child1 = (mask & p1) | ((1-mask) & p2)
        child2 = (mask & p2) | ((1-mask) & p1)
        kids = np.vstack([child1, child2])


        mut = rng.random(size=kids.shape) < mutation_rate
        kids = np.where(mut, 1-kids, kids).astype(np.int8)

     
        if max_items:
            sums = kids.sum(axis=1)
            for i, s in enumerate(sums):
                if s > max_items:
                    ones = np.where(kids[i]==1)[0]
                    drop = rng.choice(ones, size=s-max_items, replace=False)
                    kids[i, drop] = 0

        # elitism
        elite = max(1, int(pop_size*elite_frac))
        elite_ix = np.argpartition(err, elite-1)[:elite]
        pop = np.vstack([pop[elite_ix], kids])[:pop_size]

    return None
```

```python
def run_21_direct(bank_df, ledger_df):
    t0 = time.perf_counter()
    b = bank_df.copy(); l = ledger_df.copy()
    b["match_key"] = b[BANK_AMOUNT_COL].round(2).astype(str)
    l["match_key"] = l[LEDGER_AMOUNT_COL].round(2).astype(str)
    bk = b["match_key"].value_counts(); lk = l["match_key"].value_counts()
    uniq = set(bk[bk==1].index) & set(lk[lk==1].index)
    bm = b[b["match_key"].isin(uniq)]
    lm = l[l["match_key"].isin(uniq)]
    dm = bm.merge(lm, on="match_key", suffixes=("_bank","_ledger"))
    took = time.perf_counter()-t0
    stats = {"method":"2.1_direct","seconds":took,"matches_found":len(dm),"rows_processed":len(bank_df),"ms_per_row":(1000*took/max(1,len(bank_df)))}
    # tidy output
    dm_out = dm[[BANK_ID_COL, f"{LEDGER_ID_COL}"] if f"{LEDGER_ID_COL}" in dm.columns else [f"{LEDGER_ID_COL}_ledger"]]
    return dm, stats

def run_exact(bank_df, ledger_df, solver="mitm",
              abs_tol_c=500, rel_tol=0.02, max_k=24, max_items=6, ga_time=0.3):
    Ls, cents_arr, pos_by_orig = build_sorted_ledger(ledger_df)
    avail = fresh_avail(len(Ls))
    t0 = time.perf_counter()
    matches = []; totalK=0

    for _, r in bank_df.iterrows():
        a = pd.to_numeric(r[BANK_AMOUNT_COL], errors="coerce")
        if pd.isna(a) or a==0: continue
        target_c = int(round(abs(float(a))*100))

        cand = pick_window_by_amount(target_c, cents_arr, Ls, avail, abs_tol_c, rel_tol, max_k)
        if cand.empty: continue
        totalK += len(cand)
        cents = cand["cents"].to_numpy(np.int64)
        if cents.min(initial=target_c+1) > target_c: continue
        if cents.sum() < target_c: continue
        g = gcd_reduce(cents.tolist())
        if g and (target_c % g != 0): continue

        if solver=="mitm":
            chosen = mitm_subset_sum(cand, target_c)
        elif solver=="dp":
            chosen = dp_subset_sum(cand, target_c)
        elif solver=="ga":
            chosen = ga_subset_sum(cand, target_c, max_items=max_items, time_budget_sec=ga_time)
        else:
            raise ValueError("solver must be mitm|dp|ga")

        if chosen:
            block = ledger_df.loc[chosen, [LEDGER_ID_COL, LEDGER_AMOUNT_COL]]
            matches.append({
                "bank_tx_id":  r[BANK_ID_COL],
                "bank_amount": round(float(a), 2),
                "ledger_ids":  block[LEDGER_ID_COL].tolist(),
                "ledger_amts": block[LEDGER_AMOUNT_COL].round(2).tolist(),
                "sum_ledger":  round(float(block[LEDGER_AMOUNT_COL].sum()), 2),
                "n_items":     len(block),
                "method":      f"exact_{solver}"
            })
            avail[pos_by_orig[chosen].to_numpy()] = False

    took = time.perf_counter()-t0
    df = pd.DataFrame(matches)
    stats = {
        "method": f"exact_{solver}",
        "seconds": took,
        "rows_processed": len(bank_df),
        "matches_found": len(df),
        "avg_candidates": totalK/max(1,len(bank_df)),
        "ms_per_row": (1000*took/max(1,len(bank_df)))
    }
    return df, stats
```

```python
def clean_desc(s):
    if pd.isna(s): return ""
    s = str(s).lower().strip()
    s = re.sub(r"[\W_]+", " ", s)
    return re.sub(r"\s+", " ", s).strip()

def seq_ratio(a, b): return SequenceMatcher(a=a or "", b=b or "").ratio()
def jaccard(a, b):
    sa = set((a or "").split()); sb = set((b or "").split())
    return len(sa & sb)/max(1, len(sa | sb))
def trigram_bag(s):
    s = f"  {s or ''}  "; return [s[i:i+3] for i in range(max(0,len(s)-2))]
def trigram_cos(a,b):
    from collections import Counter
    ca, cb = Counter(trigram_bag(a)), Counter(trigram_bag(b))
    dot = sum(ca[g]*cb.get(g,0) for g in ca)
    na = np.sqrt(sum(v*v for v in ca.values())); nb = np.sqrt(sum(v*v for v in cb.values()))
    return float(dot)/(na*nb) if na and nb else 0.0

def run_fuzzy(bank_df, ledger_df, top_n=5, amt_abs_tol=5.0, amt_rel_tol=0.02, bucket=100.0,
              positives=None, thresh=0.7):
    t0 = time.perf_counter()
    b = bank_df.copy(); l = ledger_df.copy()
    b["amt_abs"] = pd.to_numeric(b[BANK_AMOUNT_COL], errors="coerce").abs()
    l["amt_abs"] = pd.to_numeric(l[LEDGER_AMOUNT_COL], errors="coerce").abs()
    b["desc_clean"] = b[BANK_DESC_COL].astype(str).map(clean_desc)
    l["desc_clean"] = l[LEDGER_DESC_COL].astype(str).map(clean_desc)
    b["bucket"] = (b["amt_abs"]//bucket).astype("Int64")
    l["bucket"] = (l["amt_abs"]//bucket).astype("Int64")

    pairs = (
        b[[BANK_ID_COL,"amt_abs","bucket","desc_clean"]]
          .dropna(subset=["amt_abs"])
          .merge(l[[LEDGER_ID_COL,"amt_abs","bucket","desc_clean"]]
          .dropna(subset=["amt_abs"]), on="bucket", how="inner", suffixes=("_b","_l"))
    )
    pairs["abs_diff"] = (pairs["amt_abs_b"]-pairs["amt_abs_l"]).abs()
    pairs["rel_diff"] = pairs["abs_diff"]/pairs["amt_abs_b"].clip(lower=1e-6)
    pairs = pairs[(pairs["abs_diff"]<=amt_abs_tol)|(pairs["rel_diff"]<=amt_rel_tol)].copy()

    pairs["seq_ratio"]  = [seq_ratio(a,b)  for a,b in zip(pairs["desc_clean_b"], pairs["desc_clean_l"])]
    pairs["jaccard"]    = [jaccard(a,b)    for a,b in zip(pairs["desc_clean_b"], pairs["desc_clean_l"])]
    pairs["tri_cosine"] = [trigram_cos(a,b)for a,b in zip(pairs["desc_clean_b"], pairs["desc_clean_l"])]

    amt_s = 0.5*(1 - (pairs["abs_diff"]/amt_abs_tol).clip(upper=1)) + \
            0.5*(1 - (pairs["rel_diff"]/amt_rel_tol).clip(upper=1))
    txt_s = 0.5*pairs["seq_ratio"] + 0.3*pairs["jaccard"] + 0.2*pairs["tri_cosine"]
    pairs["fuzzy_score"] = (0.6*amt_s + 0.4*txt_s).clip(0,1)

    pairs["rank_in_bank"] = pairs.groupby(BANK_ID_COL)["fuzzy_score"].rank(ascending=False, method="first")
    sug = (pairs.sort_values(["fuzzy_score"], ascending=False)
                .query("rank_in_bank <= @top_n")
                [[BANK_ID_COL, LEDGER_ID_COL, "fuzzy_score","rank_in_bank",
                  "amt_abs_b","amt_abs_l","abs_diff","seq_ratio","jaccard","tri_cosine"]]
                .reset_index(drop=True))

    took = time.perf_counter()-t0
    stats = {"method":"fuzzy","seconds":took,"rows_processed":len(bank_df),
             "ms_per_row": (1000*took/max(1,len(bank_df))), "suggestions": len(sug)}

    if positives is not None and len(positives):
        lab = sug.merge(positives.assign(label=1),
                        left_on=[BANK_ID_COL, LEDGER_ID_COL],
                        right_on=["tx_id","target_id"], how="left")
        lab["label"] = lab["label"].fillna(0).astype(int)
        top1 = (lab.sort_values(["fuzzy_score"], ascending=False)
                    .groupby(BANK_ID_COL).head(1))
        stats["top1_accuracy"] = float(top1["label"].mean()) if len(top1) else np.nan
        stats["precision_at_thresh"] = float(lab.loc[lab["fuzzy_score"]>=thresh,"label"].mean()) \
                                       if (lab["fuzzy_score"]>=thresh).any() else np.nan
    return sug, stats
```

```python
def sample_bank(n, base, seed=42):
    rng = np.random.default_rng(seed)
    n = min(n, len(base))
    return base.iloc[rng.choice(base.index.to_numpy(), size=n, replace=False)].reset_index(drop=True)

def bench_methods(sizes=(100, 250, 500, 750, 1000),
                  methods=("2.1_direct","exact_mitm","exact_dp","exact_ga","fuzzy"),
                  repeats=2,
                  params=None,
                  seed=42):
    if params is None:
        params = {
            "abs_tol_c": 500, "rel_tol": 0.02, "max_k": 24,
            "ga_max_items": 6, "ga_time": 0.3,
            "top_n": 5, "fuzzy_amt_abs_tol": 5.0, "fuzzy_amt_rel_tol": 0.02
        }
    results = []
    for n in sizes:
        for rep in range(repeats):
            bank_n = sample_bank(n, bank_base, seed=seed+rep+n)
            if "2.1_direct" in methods:
                _, st = run_21_direct(bank_n, ledger_base)
                st.update({"bank_rows": n, "repeat": rep+1})
                print(f"[2.1] N={n:4d} rep{rep+1}: {st['seconds']:.2f}s, matches={st['matches_found']}")
                results.append(st)

            if "exact_mitm" in methods:
                _, st = run_exact(bank_n, ledger_base, solver="mitm",
                                  abs_tol_c=params["abs_tol_c"], rel_tol=params["rel_tol"], max_k=params["max_k"])
                st.update({"bank_rows": n, "repeat": rep+1})
                print(f"[MITM] N={n:4d} rep{rep+1}: {st['seconds']:.2f}s, matches={st['matches_found']}, K~{st.get('avg_candidates',0):.1f}")
                results.append(st)

            if "exact_dp" in methods:
                _, st = run_exact(bank_n, ledger_base, solver="dp",
                                  abs_tol_c=params["abs_tol_c"], rel_tol=params["rel_tol"], max_k=params["max_k"])
                st.update({"bank_rows": n, "repeat": rep+1})
                print(f"[DP  ] N={n:4d} rep{rep+1}: {st['seconds']:.2f}s, matches={st['matches_found']}, K~{st.get('avg_candidates',0):.1f}")
                results.append(st)

            if "exact_ga" in methods:
                _, st = run_exact(bank_n, ledger_base, solver="ga",
                                  abs_tol_c=params["abs_tol_c"], rel_tol=params["rel_tol"], max_k=params["max_k"],
                                  max_items=params["ga_max_items"], ga_time=params["ga_time"])
                st.update({"bank_rows": n, "repeat": rep+1})
                print(f"[GA  ] N={n:4d} rep{rep+1}: {st['seconds']:.2f}s, matches={st['matches_found']}, K~{st.get('avg_candidates',0):.1f}")
                results.append(st)

            if "fuzzy" in methods:
                _, st = run_fuzzy(bank_n, ledger_base, top_n=params["top_n"],
                                  amt_abs_tol=params["fuzzy_amt_abs_tol"], amt_rel_tol=params["fuzzy_amt_rel_tol"],
                                  positives=positives if len(positives) else None)
                st.update({"bank_rows": n, "repeat": rep+1})
                acc = st.get("top1_accuracy", np.nan)
                print(f"[Fuzzy] N={n:4d} rep{rep+1}: {st['seconds']:.2f}s, top1={acc if not np.isnan(acc) else 'NA'}")
                results.append(st)

    res = pd.DataFrame(results)
    res.to_csv("task_5_1_benchmark_results.csv", index=False)
    print("Saved â†’ task_5_1_benchmark_results.csv")
    return res
```

```python
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt


if 'results' in globals() and isinstance(results, pd.DataFrame):
    res = results.copy()
else:
    res = pd.read_csv("task_5_1_benchmark_results.csv")

# 2) Basic cleanup
res = res.copy()
if "bank_rows" not in res.columns:

    res["bank_rows"] = res.get("rows_processed", np.nan)


sec = res["seconds"].replace(0, np.nan)
rows = res["bank_rows"].replace(0, np.nan)

res["ms_per_row"]   = res.get("ms_per_row", (sec / rows) * 1000.0)
res["rows_per_sec"] = rows / sec
res["match_rate"]   = res.get("matches_found", 0) / rows
res["method_clean"] = res["method"].astype(str)


outdir = "clean_outputs/bench_plots"
os.makedirs(outdir, exist_ok=True)


def line_by_method(df, xcol, ycol, title, ylabel, filename, agg="median"):
    """Line plot: y vs x for each method, aggregated over repeats."""
    plt.figure(figsize=(8,5))

    if agg == "median":
        agg_df = df.groupby(["method_clean", xcol])[ycol].median().reset_index()
    else:
        agg_df = df.groupby(["method_clean", xcol])[ycol].mean().reset_index()

    for m, sub in agg_df.groupby("method_clean"):
        sub = sub.sort_values(xcol)
        plt.plot(sub[xcol], sub[ycol], marker="o", label=m)

    plt.title(title)
    plt.xlabel(xcol)
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.legend()
    path = os.path.join(outdir, filename)
    plt.tight_layout()
    plt.savefig(path, dpi=160)
    plt.show()
    print(f"saved â†’ {path}")

def bar_by_method(df, ycol, title, ylabel, filename, agg="median"):
    plt.figure(figsize=(8,5))
    if agg == "median":
        vals = df.groupby("method_clean")[ycol].median().sort_values(ascending=False)
    else:
        vals = df.groupby("method_clean")[ycol].mean().sort_values(ascending=False)
    vals.plot(kind="bar")
    plt.title(title)
    plt.ylabel(ylabel)
    plt.grid(True, axis="y", alpha=0.3)
    path = os.path.join(outdir, filename)
    plt.tight_layout()
    plt.savefig(path, dpi=160)
    plt.show()
    print(f"saved â†’ {path}")

def scatter_xy(df, xcol, ycol, title, xlabel, ylabel, filename):
    plt.figure(figsize=(7,5))
    plt.scatter(df[xcol], df[ycol], alpha=0.6)
    plt.title(title)
    plt.xlabel(xcol); plt.ylabel(ycol)
    plt.grid(True, alpha=0.3)
    path = os.path.join(outdir, filename)
    plt.tight_layout()
    plt.savefig(path, dpi=160)
    plt.show()
    print(f"saved â†’ {path}")


line_by_method(res, "bank_rows", "seconds",
               title="Runtime vs Bank Rows",
               ylabel="Seconds (median over repeats)",
               filename="runtime_vs_size.png")


line_by_method(res, "bank_rows", "ms_per_row",
               title="Latency (ms per row) vs Bank Rows",
               ylabel="ms per row (median)",
               filename="ms_per_row_vs_size.png")


line_by_method(res, "bank_rows", "rows_per_sec",
               title="Throughput (rows/sec) vs Bank Rows",
               ylabel="Rows per second (median)",
               filename="throughput_vs_size.png")


if "matches_found" in res.columns:
    line_by_method(res, "bank_rows", "matches_found",
                   title="Matches Found vs Bank Rows",
                   ylabel="# Matches (median)",
                   filename="matches_vs_size.png")


if "matches_found" in res.columns:
    line_by_method(res, "bank_rows", "match_rate",
                   title="Match Rate vs Bank Rows",
                   ylabel="Match rate (median)",
                   filename="match_rate_vs_size.png")


if "top1_accuracy" in res.columns:
    line_by_method(res.dropna(subset=["top1_accuracy"]), "bank_rows", "top1_accuracy",
                   title="Fuzzy Top-1 Accuracy vs Bank Rows",
                   ylabel="Top-1 accuracy",
                   filename="fuzzy_top1_vs_size.png")


bar_by_method(res, "seconds",
              title="Overall Runtime by Method (median)",
              ylabel="Seconds",
              filename="bar_runtime_overall.png")

bar_by_method(res, "ms_per_row",
              title="Overall Latency by Method (median)",
              ylabel="ms per row",
              filename="bar_ms_per_row_overall.png")

if "matches_found" in res.columns:
    bar_by_method(res, "matches_found",
                  title="Overall Matches Found by Method (median)",
                  ylabel="# Matches",
                  filename="bar_matches_overall.png")


if "avg_candidates" in res.columns:
    scatter_xy(res.dropna(subset=["avg_candidates","ms_per_row"]),
               "avg_candidates", "ms_per_row",
               title="Latency vs Candidate Set Size",
               xlabel="Avg candidates per bank row (K)",
               ylabel="ms per row",
               filename="scatter_latency_vs_K.png")


def med(x): return np.nanmedian(x)
summary_by_method = (
    res.groupby("method_clean")
      .agg(
          runs=("method_clean","size"),
          bank_rows=("bank_rows","sum"),
          seconds=("seconds", med),
          ms_per_row=("ms_per_row", med),
          rows_per_sec=("rows_per_sec", med),
          matches_found=("matches_found", med) if "matches_found" in res.columns else ("method_clean","size"),
          top1_acc=("top1_accuracy", med) if "top1_accuracy" in res.columns else ("method_clean","size")
      )
      .reset_index()
)
print("\n=== Summary by method (median across repeats/sizes) ===")
display(summary_by_method)

summary_by_size = (
    res.groupby(["method_clean","bank_rows"])
      .agg(
          seconds=("seconds", med),
          ms_per_row=("ms_per_row", med),
          rows_per_sec=("rows_per_sec", med),
          matches_found=("matches_found", med) if "matches_found" in res.columns else ("method_clean","size"),
      )
      .reset_index()
)
print("\n=== Summary by method & size (medians) ===")
display(summary_by_size.head(20))

# Save tables
summary_by_method.to_csv(os.path.join(outdir, "summary_by_method.csv"), index=False)
summary_by_size.to_csv(os.path.join(outdir, "summary_by_size.csv"), index=False)
print(f"\nSaved tables â†’ {outdir}/summary_by_method.csv and summary_by_size.csv")
```
